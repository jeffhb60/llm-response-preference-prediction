# DeBERTa Cross-Encoder Preference Classifier

This notebook builds a transformer-based cross-encoder for the Kaggle **LLM Classification Finetuning** competition.

**Project Goal**

Given `prompt`, `response_a`, and `response_b`, predict probabilities for:

- `winner_model_a`
- `winner_model_b`
- `winner_tie`

**Approach** 

1. Load train/test data
2. Parse JSON-like text fields
3. Create 3-class targets
4. Build pairwise comparison text
5. Add A/B swap augmentation
6. Fine-tune a DeBERTa sequence classifier
7. Validate with multiclass log loss
8. Train final model
9. Generate `submission.csv`

## 01. Setup and Config 

```python
import os
import json
import gc
import math
import random
import warnings
from pathlib import Path

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    log_loss,
    accuracy_score,
    classification_report,
    confusion_matrix,
)

import torch
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW

from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    get_linear_schedule_with_warmup,
)
```

```python
warnings.filterwarnings("ignore")
RANDOM_STATE = 42
```

```python
def seed_everything(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

seed_everything(RANDOM_STATE)
```

```python
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("Device:", DEVICE)
print("Torch version:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())

if torch.cuda.is_available():
    print("CUDA device count:", torch.cuda.device_count())
    for i in range(torch.cuda.device_count()):
        print(i, torch.cuda.get_device_name(i))
    print("Active GPU:", torch.cuda.get_device_name(0))
    print("CUDA device capability:", torch.cuda.get_device_capability(0))
    print("PyTorch CUDA version:", torch.version.cuda)
```

```python
# Set to 2000 for debug.
# Set to None for a real run.
DEBUG_N_ROWS = None
# DEBUG_N_ROWS = 2000

MAX_LENGTH = 512

BATCH_SIZE = 2
EVAL_BATCH_SIZE = 4
GRAD_ACCUM_STEPS = 4
EPOCHS = 2

# Safer learning rate based on earlier NaN issues.
LEARNING_RATE = 2e-6

WEIGHT_DECAY = 0.01
WARMUP_RATIO = 0.06
MAX_GRAD_NORM = 1.0

VALID_SIZE = 0.15

USE_SWAP_AUGMENTATION = True

# Keep off for stability.
USE_AMP = False

RUN_FINAL_FULL_DATA_TRAINING = False

print({
    "DEBUG_N_ROWS": DEBUG_N_ROWS,
    "MAX_LENGTH": MAX_LENGTH,
    "BATCH_SIZE": BATCH_SIZE,
    "EVAL_BATCH_SIZE": EVAL_BATCH_SIZE,
    "GRAD_ACCUM_STEPS": GRAD_ACCUM_STEPS,
    "EPOCHS": EPOCHS,
    "LEARNING_RATE": LEARNING_RATE,
    "USE_AMP": USE_AMP,
    "USE_SWAP_AUGMENTATION": USE_SWAP_AUGMENTATION,
    "RUN_FINAL_FULL_DATA_TRAINING": RUN_FINAL_FULL_DATA_TRAINING,
})
```

```python
DATA_DIR = Path("/kaggle/input/competitions/llm-classification-finetuning")

TRAIN_PATH = DATA_DIR / "train.csv"
TEST_PATH = DATA_DIR / "test.csv"
SAMPLE_SUB_PATH = DATA_DIR / "sample_submission.csv"

print("Using input directory:", DATA_DIR)
print("Train path:", TRAIN_PATH)
print("Test path:", TEST_PATH)
print("Sample submission path:", SAMPLE_SUB_PATH)
```

```python
MODEL_PATH = '/kaggle/input/models/mtnash/deberta-v3-small/transformers/default/1'
print("Using model path:", MODEL_PATH)
print("Files in model path:")
import os
for dirname, _, filenames in os.walk(MODEL_PATH):
    for filename in filenames:
        print(filename)
```

## 02. Load Data 

```python
train = pd.read_csv(TRAIN_PATH)
test = pd.read_csv(TEST_PATH)
sample_submission = pd.read_csv(SAMPLE_SUB_PATH)

if DEBUG_N_ROWS is not None:
    train = train.sample(DEBUG_N_ROWS, random_state=RANDOM_STATE).reset_index(drop=True)

print("Train shape:", train.shape)
print("Test shape:", test.shape)
print("Sample submission shape:", sample_submission.shape)

display(train.head())
display(test.head())
display(sample_submission.head())
```

## 03. Column Checks

```python
required_train_cols = {
    "id",
    "prompt",
    "response_a",
    "response_b",
    "winner_model_a",
    "winner_model_b",
    "winner_tie",
}

required_test_cols = {
    "id",
    "prompt",
    "response_a",
    "response_b",
}

missing_train = required_train_cols - set(train.columns)
missing_test = required_test_cols - set(test.columns)

if missing_train:
    raise ValueError(f"Missing required train columns: {missing_train}")

if missing_test:
    raise ValueError(f"Missing required test columns: {missing_test}")

print("Column checks passed.")
```

## 04. Parse Text Fields 

```python
def parse_text_field(x):
    """
    Safely parse text fields that may be JSON-encoded lists.

    Examples:
    '["hello", "world"]' -> 'hello\\nworld'
    plain text -> plain text
    NaN -> ''
    """
    if pd.isna(x):
        return ""

    if not isinstance(x, str):
        return str(x)

    try:
        value = json.loads(x)

        if isinstance(value, list):
            return "\n".join("" if item is None else str(item) for item in value)

        if value is None:
            return ""

        return str(value)

    except Exception:
        return x


for col in ["prompt", "response_a", "response_b"]:
    train[f"{col}_parsed"] = train[col].apply(parse_text_field)
    test[f"{col}_parsed"] = test[col].apply(parse_text_field)

display(train[["prompt_parsed", "response_a_parsed", "response_b_parsed"]].head(2))
```

## 05. Target Labels 

```python
target_cols = ["winner_model_a", "winner_model_b", "winner_tie"]

train["target"] = train[target_cols].values.argmax(axis=1)

label_names = {
    0: "winner_model_a",
    1: "winner_model_b",
    2: "winner_tie",
}

print("Target distribution:")
target_dist = train["target"].value_counts(normalize=True).sort_index()

for idx, value in target_dist.items():
    print(f"{idx} ({label_names[idx]}): {value:.4f}")

display(train[target_cols + ["target"]].head())
```

## 06. Build Cross-Encoder Text 

```python
def build_comparison_text(df):
    return (
        "Prompt:\n" + df["prompt_parsed"].fillna("") +
        "\n\nResponse A:\n" + df["response_a_parsed"].fillna("") +
        "\n\nResponse B:\n" + df["response_b_parsed"].fillna("")
    )


train["text"] = build_comparison_text(train)
test["text"] = build_comparison_text(test)

print(train["text"].iloc[0][:2000])
```

## 07. A/B Swap Augmentation

```python
def create_swapped_training_data(df):
    """
    Create augmented rows where response_a and response_b are swapped.

    Label mapping:
    0 -> 1
    1 -> 0
    2 -> 2
    """
    swapped = df.copy()

    swapped["response_a_parsed"] = df["response_b_parsed"].values
    swapped["response_b_parsed"] = df["response_a_parsed"].values

    swapped["text"] = build_comparison_text(swapped)

    swapped["target"] = df["target"].map({
        0: 1,
        1: 0,
        2: 2,
    }).astype(int).values

    return swapped


swapped_preview = create_swapped_training_data(train.head(3))
print("Original targets:", train.loc[:2, "target"].tolist())
print("Swapped targets:", swapped_preview["target"].tolist())
```

## 08. Train/Validation Split 

```python
train_base, valid_base = train_test_split(
    train,
    test_size=VALID_SIZE,
    random_state=RANDOM_STATE,
    stratify=train["target"],
)

if USE_SWAP_AUGMENTATION:
    train_base_swapped = create_swapped_training_data(train_base)

    train_model_df = pd.concat(
        [
            train_base[["text", "target"]],
            train_base_swapped[["text", "target"]],
        ],
        axis=0,
        ignore_index=True,
    )
else:
    train_model_df = train_base[["text", "target"]].copy()

valid_model_df = valid_base[["text", "target"]].copy()

print("Train rows before augmentation:", train_base.shape[0])
print("Train rows after augmentation:", train_model_df.shape[0])
print("Validation rows:", valid_model_df.shape[0])

print("\nTrain target distribution:")
print(train_model_df["target"].value_counts(normalize=True).sort_index())

print("\nValidation target distribution:")
print(valid_model_df["target"].value_counts(normalize=True).sort_index())
```

## 09. Tokenizer and Dataset 

```python
tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH, use_fast=True)

print("Tokenizer loaded.")
print("Tokenizer model max length:", tokenizer.model_max_length)


class PreferenceDataset(Dataset):
    def __init__(self, texts, labels=None, tokenizer=None, max_length=256):
        self.texts = list(texts)
        self.labels = None if labels is None else list(labels)
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]

        encoded = self.tokenizer(
            text,
            truncation=True,
            max_length=self.max_length,
            padding=False,
            return_tensors=None,
        )

        item = {
            "input_ids": torch.tensor(encoded["input_ids"], dtype=torch.long),
            "attention_mask": torch.tensor(encoded["attention_mask"], dtype=torch.long),
        }

        if self.labels is not None:
            item["labels"] = torch.tensor(self.labels[idx], dtype=torch.long)

        return item


def collate_fn(batch):
    input_ids = [x["input_ids"] for x in batch]
    attention_mask = [x["attention_mask"] for x in batch]

    padded = tokenizer.pad(
        {
            "input_ids": input_ids,
            "attention_mask": attention_mask,
        },
        padding=True,
        return_tensors="pt",
    )

    if "labels" in batch[0]:
        padded["labels"] = torch.stack([x["labels"] for x in batch])

    return padded
```

## 10. DataLoaders 

```python
train_dataset = PreferenceDataset(
    texts=train_model_df["text"].values,
    labels=train_model_df["target"].values,
    tokenizer=tokenizer,
    max_length=MAX_LENGTH,
)

valid_dataset = PreferenceDataset(
    texts=valid_model_df["text"].values,
    labels=valid_model_df["target"].values,
    tokenizer=tokenizer,
    max_length=MAX_LENGTH,
)

train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
    collate_fn=collate_fn,
    num_workers=2,
    pin_memory=True,
)

valid_loader = DataLoader(
    valid_dataset,
    batch_size=EVAL_BATCH_SIZE,
    shuffle=False,
    collate_fn=collate_fn,
    num_workers=2,
    pin_memory=True,
)

print("Train batches:", len(train_loader))
print("Validation batches:", len(valid_loader))
```

## 11. Model, Optimizer, and Scheduler 

```python
model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_PATH,
    num_labels=3,
    torch_dtype=torch.float32,
)

model.to(DEVICE)

optimizer = AdamW(
    model.parameters(),
    lr=LEARNING_RATE,
    weight_decay=WEIGHT_DECAY,
)

num_update_steps_per_epoch = math.ceil(len(train_loader) / GRAD_ACCUM_STEPS)
num_training_steps = EPOCHS * num_update_steps_per_epoch
num_warmup_steps = int(WARMUP_RATIO * num_training_steps)

scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=num_warmup_steps,
    num_training_steps=num_training_steps,
)

scaler = torch.cuda.amp.GradScaler(enabled=False)

print("Training steps:", num_training_steps)
print("Warmup steps:", num_warmup_steps)
print("Model dtype:", next(model.parameters()).dtype)
```

## 12. Helper Functions 

```python
def move_batch_to_device(batch, device):
    return {k: v.to(device) for k, v in batch.items()}


def train_one_epoch(model, loader, optimizer, scheduler, scaler=None):
    model.train()

    total_loss = 0.0
    optimizer.zero_grad(set_to_none=True)

    use_amp_now = USE_AMP and DEVICE.type == "cuda"

    for step, batch in enumerate(loader):
        batch = move_batch_to_device(batch, DEVICE)

        if use_amp_now:
            with torch.cuda.amp.autocast():
                outputs = model(**batch)
                loss_raw = outputs.loss

                if torch.isnan(loss_raw) or torch.isinf(loss_raw):
                    print(f"Bad loss at step {step + 1}: {loss_raw}")
                    print("Labels:", batch["labels"][:10])
                    print("Input IDs min/max:", batch["input_ids"].min().item(), batch["input_ids"].max().item())
                    raise ValueError("NaN or Inf loss detected.")

                loss = loss_raw / GRAD_ACCUM_STEPS

            scaler.scale(loss).backward()

            if (step + 1) % GRAD_ACCUM_STEPS == 0 or (step + 1) == len(loader):
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), MAX_GRAD_NORM)

                scaler.step(optimizer)
                scaler.update()

                optimizer.zero_grad(set_to_none=True)
                scheduler.step()

        else:
            outputs = model(**batch)
            loss_raw = outputs.loss

            if torch.isnan(loss_raw) or torch.isinf(loss_raw):
                print(f"Bad loss at step {step + 1}: {loss_raw}")
                print("Labels:", batch["labels"][:10])
                print("Input IDs min/max:", batch["input_ids"].min().item(), batch["input_ids"].max().item())
                raise ValueError("NaN or Inf loss detected.")

            loss = loss_raw / GRAD_ACCUM_STEPS
            loss.backward()

            if (step + 1) % GRAD_ACCUM_STEPS == 0 or (step + 1) == len(loader):
                torch.nn.utils.clip_grad_norm_(model.parameters(), MAX_GRAD_NORM)

                optimizer.step()
                optimizer.zero_grad(set_to_none=True)
                scheduler.step()

        total_loss += loss_raw.item()

        if (step + 1) % 25 == 0:
            print(f"Step {step + 1}/{len(loader)} | loss: {total_loss / (step + 1):.5f}")

    return total_loss / len(loader)


@torch.no_grad()
def predict_proba(model, loader):
    model.eval()

    all_probs = []
    all_labels = []

    for batch in loader:
        labels = batch.get("labels")

        batch = move_batch_to_device(batch, DEVICE)

        outputs = model(
            input_ids=batch["input_ids"],
            attention_mask=batch["attention_mask"],
        )

        logits = outputs.logits.detach().float().cpu().numpy()

        exp_logits = np.exp(logits - logits.max(axis=1, keepdims=True))
        probs = exp_logits / exp_logits.sum(axis=1, keepdims=True)

        all_probs.append(probs)

        if labels is not None:
            all_labels.append(labels.cpu().numpy())

    all_probs = np.vstack(all_probs)

    if all_labels:
        all_labels = np.concatenate(all_labels)
        return all_probs, all_labels

    return all_probs, None


def clip_and_normalize(preds, eps=1e-5):
    preds = np.clip(preds, eps, 1 - eps)
    return preds / preds.sum(axis=1, keepdims=True)
```

## 13. Sanity Check, Initial Batch 

```python
batch = next(iter(train_loader))
batch = move_batch_to_device(batch, DEVICE)

with torch.no_grad():
    outputs = model(**batch)

print("Initial loss:", outputs.loss)
print("Any NaN in logits:", torch.isnan(outputs.logits).any().item())
print("Logits:", outputs.logits[:5])
print("Labels:", batch["labels"][:10])
print("Input IDs min/max:", batch["input_ids"].min().item(), batch["input_ids"].max().item())
```

## 14. Train Validation Model 

```python
best_valid_logloss = float("inf")
best_state_dict = None

for epoch in range(EPOCHS):
    print("\n" + "=" * 80)
    print(f"Epoch {epoch + 1}/{EPOCHS}")
    print("=" * 80)

    train_loss = train_one_epoch(
        model=model,
        loader=train_loader,
        optimizer=optimizer,
        scheduler=scheduler,
        scaler=scaler,
    )

    valid_probs, valid_labels = predict_proba(model, valid_loader)
    valid_probs = clip_and_normalize(valid_probs)

    valid_logloss = log_loss(valid_labels, valid_probs, labels=[0, 1, 2])
    valid_preds = valid_probs.argmax(axis=1)
    valid_acc = accuracy_score(valid_labels, valid_preds)

    print(f"Train loss: {train_loss:.6f}")
    print(f"Validation log loss: {valid_logloss:.6f}")
    print(f"Validation accuracy: {valid_acc:.6f}")

    if valid_logloss < best_valid_logloss:
        best_valid_logloss = valid_logloss
        best_state_dict = {
            k: v.detach().cpu().clone()
            for k, v in model.state_dict().items()
        }
        print("New best model saved in memory.")

print("\nBest validation log loss:", best_valid_logloss)

if best_state_dict is not None:
    model.load_state_dict(best_state_dict)
    model.to(DEVICE)
```

## 15. Validation Diagnostics 

```python
valid_probs, valid_labels = predict_proba(model, valid_loader)
valid_probs = clip_and_normalize(valid_probs)
valid_preds = valid_probs.argmax(axis=1)

valid_logloss = log_loss(valid_labels, valid_probs, labels=[0, 1, 2])
valid_acc = accuracy_score(valid_labels, valid_preds)

print(f"Validation Log Loss: {valid_logloss:.6f}")
print(f"Validation Accuracy: {valid_acc:.6f}")

print("\nClassification Report:")
print(classification_report(
    valid_labels,
    valid_preds,
    target_names=["winner_model_a", "winner_model_b", "winner_tie"],
    digits=4,
))

print("\nConfusion Matrix:")
print(confusion_matrix(valid_labels, valid_preds))

prob_df = pd.DataFrame(
    valid_probs,
    columns=["winner_model_a", "winner_model_b", "winner_tie"],
)

print("\nAverage predicted probabilities:")
display(prob_df.mean())

print("\nActual validation class distribution:")
display(pd.Series(valid_labels).value_counts(normalize=True).sort_index().rename(index=label_names))
```

## 16. Prior Blending 

```python
class_priors = train_base[["winner_model_a", "winner_model_b", "winner_tie"]].mean().values

print("Class priors from training split:")
print(class_priors)

blend_results = []

for blend_weight in [0.3, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0]:
    blended_valid_probs = (
        blend_weight * valid_probs +
        (1 - blend_weight) * class_priors
    )

    blended_valid_probs = clip_and_normalize(blended_valid_probs)
    score = log_loss(valid_labels, blended_valid_probs, labels=[0, 1, 2])

    blend_results.append({
        "blend_weight": blend_weight,
        "log_loss": score,
    })

blend_results_df = pd.DataFrame(blend_results)
display(blend_results_df)

best_row = blend_results_df.loc[blend_results_df["log_loss"].idxmin()]

BEST_BLEND_WEIGHT = float(best_row["blend_weight"])
BEST_BLEND_LOGLOSS = float(best_row["log_loss"])

print(f"Best blend weight: {BEST_BLEND_WEIGHT}")
print(f"Best blended validation log loss: {BEST_BLEND_LOGLOSS:.6f}")
```

## 17. Train Final Model on Full Data

```python
if RUN_FINAL_FULL_DATA_TRAINING:
    print("Running final full-data training.")

    del model
    gc.collect()

    if DEVICE.type == "cuda":
        torch.cuda.empty_cache()

    if USE_SWAP_AUGMENTATION:
        full_swapped = create_swapped_training_data(train)

        full_train_df = pd.concat(
            [
                train[["text", "target"]],
                full_swapped[["text", "target"]],
            ],
            axis=0,
            ignore_index=True,
        )
    else:
        full_train_df = train[["text", "target"]].copy()

    print("Full training rows:", full_train_df.shape[0])

    full_dataset = PreferenceDataset(
        texts=full_train_df["text"].values,
        labels=full_train_df["target"].values,
        tokenizer=tokenizer,
        max_length=MAX_LENGTH,
    )

    full_loader = DataLoader(
        full_dataset,
        batch_size=BATCH_SIZE,
        shuffle=True,
        collate_fn=collate_fn,
        num_workers=2,
        pin_memory=True,
    )

    final_model = AutoModelForSequenceClassification.from_pretrained(
        MODEL_PATH,
        num_labels=3,
        torch_dtype=torch.float32,
    )

    final_model.to(DEVICE)

    final_optimizer = AdamW(
        final_model.parameters(),
        lr=LEARNING_RATE,
        weight_decay=WEIGHT_DECAY,
    )

    final_num_update_steps_per_epoch = math.ceil(len(full_loader) / GRAD_ACCUM_STEPS)
    final_num_training_steps = EPOCHS * final_num_update_steps_per_epoch
    final_num_warmup_steps = int(WARMUP_RATIO * final_num_training_steps)

    final_scheduler = get_linear_schedule_with_warmup(
        final_optimizer,
        num_warmup_steps=final_num_warmup_steps,
        num_training_steps=final_num_training_steps,
    )

    final_scaler = torch.cuda.amp.GradScaler(enabled=False)

    print("Final training steps:", final_num_training_steps)
    print("Final warmup steps:", final_num_warmup_steps)

    for epoch in range(EPOCHS):
        print("\n" + "=" * 80)
        print(f"Final training epoch {epoch + 1}/{EPOCHS}")
        print("=" * 80)

        final_train_loss = train_one_epoch(
            model=final_model,
            loader=full_loader,
            optimizer=final_optimizer,
            scheduler=final_scheduler,
            scaler=final_scaler,
        )

        print(f"Final train loss: {final_train_loss:.6f}")

    inference_model = final_model

else:
    print("Skipping final full-data training.")
    print("Using validation-trained model for inference.")
    inference_model = model
```

## 18. Build Test DataLoader

```python
test_dataset = PreferenceDataset(
    texts=test["text"].values,
    labels=None,
    tokenizer=tokenizer,
    max_length=MAX_LENGTH,
)

test_loader = DataLoader(
    test_dataset,
    batch_size=EVAL_BATCH_SIZE,
    shuffle=False,
    collate_fn=collate_fn,
    num_workers=2,
    pin_memory=True,
)

print("Test batches:", len(test_loader))
```

## 19. Predict Test Set 

```python
test_probs_raw, _ = predict_proba(inference_model, test_loader)
test_probs_raw = clip_and_normalize(test_probs_raw)

print("Test probability shape:", test_probs_raw.shape)
print(test_probs_raw[:5])
print("Row sums:", test_probs_raw.sum(axis=1)[:5])
```

## 20. Apply Final Prior Blending 

```python
full_class_priors = train[["winner_model_a", "winner_model_b", "winner_tie"]].mean().values

print("Full training class priors:")
print(full_class_priors)

test_probs = (
    BEST_BLEND_WEIGHT * test_probs_raw +
    (1 - BEST_BLEND_WEIGHT) * full_class_priors
)

test_probs = clip_and_normalize(test_probs)

print("First test rows after final blending:")
print(test_probs[:5])

print("\nRow sums:")
print(test_probs.sum(axis=1)[:5])
```

## 21. Create and Save Submission 

```python
submission = pd.DataFrame({
    "id": test["id"],
    "winner_model_a": test_probs[:, 0],
    "winner_model_b": test_probs[:, 1],
    "winner_tie": test_probs[:, 2],
})

submission = submission[[
    "id",
    "winner_model_a",
    "winner_model_b",
    "winner_tie",
]]

display(submission.head())

print("Submission shape:", submission.shape)

print("\nProbability row sums:")
display(submission[["winner_model_a", "winner_model_b", "winner_tie"]].sum(axis=1).describe())
```

```python
submission.to_csv("submission.csv", index=False)

print("Saved submission.csv")
display(pd.read_csv("submission.csv").head())
```
